```
import pandas as pd
import numpy as np
import re
```

```
pd.set_option('max_columns',None)
```

```
data=pd.read_csv(path+'raw_data.csv')
```

```
data.head()
```

```
# create college code column
def col_code(x):
  l=re.findall(r'^\d{4}',x)
  if len(l)!=0:
    return l[0]
  else:
    return np.nan

data['College_Code']=data['0'].map(col_code)
data['College_Code'].fillna(axis=0,method='ffill',inplace=True) 
```

```
# create branch code column
def branch_code(x):
  l=re.findall(r'\d{9}[\w]?',x)
  if len(l)!=0:
    return l[0][4:]
  else:
    return np.nan

data['Branch_Code']=data['0'].map(branch_code)
data['Branch_Code'].fillna(axis=0,method='ffill',inplace=True)
```

```
# create college name column
def college_name(x):
  l=re.findall(r'^\d{4} ',x)
  if len(l)!=0:
    n=x.find('\n')
    return x[7:n]
  else:
    return np.nan

data['College_Name']=data['0'].map(college_name)

# remove text from entry
def rm_college_name(x):
  l=re.findall(r'^\d{4} ',x)
  if len(l)!=0:
    n=x.find('\n')
    return x.replace(x[:n+1],'')
  else:
    return x

data['0']=data['0'].map(rm_college_name)
data['College_Name'].fillna(axis=0,method='ffill',inplace=True)
```

```
# create branch name column
def branch_name(x):
  l=re.findall(r'^\d{9}',x)
  if len(l)!=0:
    n=x.find('-')
    m=x.find('\n')
    return x[n+2:m]
  else:
    return np.nan

data['Branch_Name']=data['0'].map(branch_name)

# remove text from entry
def rm_branch_name(x):
  l=re.findall(r'^\d{9}',x)
  if len(l)!=0:
    n=x.find('\n')
    return x.replace(x[:n+1],'')
  else:
    return x

data['0']=data['0'].map(rm_branch_name)
data['Branch_Name'].fillna(axis=0,method='ffill',inplace=True)
```

```
# create college type column
def college_type(x):
  l=re.findall(r'^Status:',x)
  if len(l)!=0:
    n=x.find('Home')
    if len(x[11:n-1])>0:
      return x[11:n-1]
    else:
      return 'none'
  else:
    return np.nan

data['College_Type']=data['0'].map(college_type)

# remove text from entry
def rm_college_type(x):
  l=re.findall(r'^Status:',x)
  if len(l)!=0:
    n=x.find('Home')
    return x.replace(x[:n],'')
  else:
    return x

data['0']=data['0'].map(rm_college_type)
data['College_Type'].fillna(axis=0,method='ffill',inplace=True)
```

```
# create university name column
def university_name(x):
  l=re.findall(r'^Home',x)
  if len(l)!=0:
    n=x.find(':')
    m=x.find('\n')
    return x[n+2:m]
  else:
    return np.nan

data['University_Name']=data['0'].map(university_name)

# remove text from entry
def rm_university_name(x):
  l=re.findall(r'^Home',x)
  if len(l)!=0:
    n=x.find(':')
    m=x.find('\n')
    return x.replace(x[:m+1],'')
  else:
    return x

data['0']=data['0'].map(rm_university_name)
data['University_Name'].fillna(axis=0,method='ffill',inplace=True)
```

```
# create seat type column
def seat_type(x):
  if len(x)>2:
    return x
  else:
    return np.nan

data['Seat_Type']=data['0'].map(seat_type)
data['Seat_Type'].fillna(axis=0,method='ffill',inplace=True)
```

```
# create stage column
def stage(x):
  if len(x)<3:
    return x
  else:
    return np.nan

data['Stage']=data['0'].map(stage)
data['Stage'].fillna(axis=0,method='bfill',inplace=True)
```

```
# all categories
categories_code={
    'LVJS', 'LNT1H', 'GSTO', 'DEFNT3S', 'PWDNT3S', 'GNT2S', 'LSTH',
    'PWDSCS', 'LSEBCH', 'GVJS', 'GSEBCS', 'PWDSTS', 'GSTS', 'EWS',
    'DEFSCS', 'LNT3O', 'PWDSEBCH', 'LOBCO', 'PWDOPENS', 'GSEBCH',
    'GOPENO', 'GNT3O', 'TFWS', 'GVJO', 'LNT3H', 'LNT2O', 'MI',
    'LOBCS', 'GSCH', 'DEFVJS', 'ORPHAN', 'LSEBCS', 'DEFSTS',
    'GNT1O', 'GNT1S', 'LSEBCO', 'LNT1O', 'LOPENH', 'PWDSCH',
    'GSCS', 'LOPENS', 'LSTS', 'GSCO', 'LNT2H', 'PWDOPENH',
    'GNT3H', 'GOBCS', 'GSEBCO', 'PWDOBCH', 'PWDNT1S', 'LNT3S',
    'GNT2O', 'LOPENO', 'DEFSEBCS', 'LVJO', 'GOPENS', 'GOBCO',
    'PWDNT2S', 'LVJH', 'LNT2S', 'LNT1S', 'DEFOPENS', 'GNT1H',
    'PWDNT3H', 'GOBCH', 'GNT2H', 'LOBCH', 'LSCS', 'LSCO', 'GSTH',
    'GNT3S', 'GOPENH', 'LSCH', 'LSTO', 'DEFOBCS', 'DEFNT2S',
    'PWDVJH', 'PWDOBCS', 'PWDSEBCS', 'GVJH'
    }
```

```
# create categories column
for i in categories_code:
  data[i]=0
```

```
# add data to respective category column
for i in range(len(data)):
  for j in range(1,24):
    if data.iloc[i,24]=='data':
      break
    else:
      if data.iloc[i,j] in (categories_code):
        data[data.iloc[i,j]][i]=data.iloc[i+1,j]
```

```
# delete data entries rows from dataframe
data=data[data['Cat_Header']!='data']
```

```
# Select only new created columns
data=data.iloc[:,25:]
```

```
# create functions for seperating rank and marks
def rank(x):
  if len(str(x))==1:
    return 0
  else:
    n=str(x).find('\n')
    return str(x)[:n]

def mark(x):
  if len(str(x))==1:
    return 0
  else:
    n=str(x).find('\n')
    return str(x)[n+2:-1]
```

```
# create seperate columns for rank and mark for category wise
for i in categories_code:
  data[i+'_rank']=data[i].map(rank)
  data[i+'_mark']=data[i].map(mark)
```

```
# drop categories columns
for i in categories_code:
  data.drop(columns=i,inplace=True)
```
