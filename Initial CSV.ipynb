```
data=pd.ExcelFile(path+'college_cutoffs.xlsx')
```

```
result=pd.DataFrame()
```

```
# concat all sheets to one dataframe
for i in data.sheet_names:
    df=pd.read_excel(data,sheet_name=i,header=None)
    result=pd.concat([result,df],axis=0)
```

```
# drop 24 index column
result.drop(columns=24,inplace=True)
```

```
# convert category name to Cat_Head entry, rank to data entry and remaining to no-data entry
def find_cat(x):
  if str(x).isupper():
    return('Cat_Head')
  elif str(x)[:2].isdigit():
    return('data')
  else:
    return('no-data')
```

```
# store result in new Cat_Header Column
result['Cat_Header']=result[1].map(find_cat)
```

```
# drop rows which contain all null values 
result.dropna(axis=0,how='all',inplace=True)
```

```
# fill null values with previous presented values 
result[0].fillna(axis=0,method='ffill',inplace=True)
```

```
# drop no-data entry values from dataframe
result=result[result['Cat_Header']!='no-data']
```

```
result.reset_index(drop=True,inplace=True)
```

some data move to next row. manually handle this data from csv file.

```
temp=result[result[0]!='I']
temp=temp[temp[0]!='II']
```

```
cat_name=set()
for i in range(len(temp)):
  for j in range(len(temp.columns)):
    if str(temp.iloc[i,j]).isupper():
      cat_name.add(str(temp.iloc[i,j]))
print(cat_name)
```
