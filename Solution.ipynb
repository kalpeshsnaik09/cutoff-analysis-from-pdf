# Task 1
Write a Python script that finds the top 10 colleges for Civil Engineering Branch based on cutoff
ranks in GOPENS/GOPENH categories for Savitribai Phule Pune University?

```
task_data=data[data['University_Name']=='Savitribai Phule Pune University']
```

```
task_data_civil=task_data[task_data['Branch_Name']=='Civil Engineering']
```

```
output=task_data_civil[task_data_civil['GOPENH_rank']!=0].sort_values(by='GOPENH_rank').iloc[:10]
```

## Solution 
Top 10 Colleges with Rank and Marks

```
output[['College_Code','College_Name','College_Type','Seat_Type','GOPENH_rank','GOPENH_mark']]
```
